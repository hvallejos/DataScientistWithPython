## Gráfico lineal(1) 

```python
#year y pop ya disponibles en el curso
import pandas as pd
mlb = pd.read_csv("./CrecimientoPob.csv", sep= ";")

year_l = mlb["Year"].tolist()
population_l = mlb["Population"].tolist()

import numpy as np
year = np.array(year_l)
pop = np.array(population_l)

# Imprime el último item de year y pop
print(year[-1])
print(pop[-1])

# Importar matplotlib.pyplot como plt
import matplotlib.pyplot as plt

# Crea un gráfico lineal donde year sea  el eje x y pop sea el eje y
plt.plot(year, pop)

# Muestra el gráfico
plt.show()
```

## Gráfico lineal(2)

```python
# Importamos los datos a probar
import pandas as pd
df = pd.read_csv('http://assets.datacamp.com/course/intermediate_python/gapminder.csv', index_col = 0)
gdp_cap = list(df.gdp_cap)
life_exp = list(df.life_exp)

# Print the last item of gdp_cap(PBI) and life_exp (esperanza de vida)
print(gdp_cap[-1])
print(life_exp[-1])

# Make a line plot, gdp_cap on the x-axis, life_exp on the y-axis
plt.plot(gdp_cap, life_exp)

# Display the plot
plt.show()
```

## Gráfico de dispersión(1)

```python
# Creando gráfico de dispersión
plt.scatter(gdp_cap, life_exp)

# Colocar el eje x en una escala logarítmica
plt.xscale('log')

# Muestra el gráfico
plt.show()
```

## Histogramas

```python
# Construyendo un histograma

# hist(lista,num de bins) si no coloco número de bins, por defecto es 10
plt.hist(life_exp)

# Muestro el histograma
plt.show()
```

## Personalizando gráficos
### Etiquetas

```python
# Basic scatter plot, log scale
plt.scatter(gdp_cap, life_exp)
plt.xscale('log') 

# Strings
xlab = 'GDP per Capita [in USD]'
ylab = 'Life Expectancy [in years]'
title = 'World Development in 2007'

# Add axis labels
plt.xlabel(xlab)
plt.ylabel(ylab)

# Add title
plt.title(title)

# After customizing, display the plot
plt.show()
```

### Ticks

```python
# Scatter plot
plt.scatter(gdp_cap, life_exp)

# Previous customizations
plt.xscale('log') 
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')

# Definition of tick_val and tick_lab
tick_val = [1000,10000,100000]
tick_lab = ['1k','10k','100k']

# Adapt the ticks on the x-axis
plt.xticks(tick_val,tick_lab)

# After customizing, display the plot
plt.show()
```
